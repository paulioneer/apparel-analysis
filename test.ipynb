```python
def find_first_and_second_numbers(input_str):
    # 정규표현식 패턴
    pattern = r'\d+'  # 숫자에 대한 패턴, \d는 숫자를 의미하고 +는 하나 이상의 숫자를 의미

    # 문자열에서 정규표현식에 해당하는 부분을 찾음
    numbers = re.findall(pattern, input_str)

    # 찾은 숫자가 두 개 이상일 경우
    if len(numbers) >= 2:
        first_number = int(numbers[0])  # 첫 번째 숫자
        second_number = int(numbers[1])  # 두 번째 숫자
        return first_number, second_number
    else:
        return None, None 

def extract_number_from_string(s):
    # 문자열에서 숫자 추출
    numbers = re.findall(r'\d+', s)
    if numbers:
        return int(numbers[0])  # 추출된 숫자가 있으면 첫 번째 숫자 반환
    else:
        return None  # 추출된 숫자가 없으면 None 반환
    
    
def parse_date(date_str):
    # 각각의 형식에 따라 날짜를 파싱
    try:
        # YYYY.MM 형식
        return datetime.strptime(date_str, '%Y.%m')
    except ValueError:
        pass

    try:
        # YY.MM 형식
        return datetime.strptime(date_str, '%y.%m')
    except ValueError:
        pass

    try:
        # YYYY.M 형식
        return datetime.strptime(date_str, '%Y.%m')
    except ValueError:
        pass

    try:
        # YY.M 형식
        return datetime.strptime(date_str, '%y.%m')
    except ValueError:
        pass

    try:
        # YYYY-MM 형식
        return datetime.strptime(date_str, '%Y-%m')
    except ValueError:
        pass

    try:
        # YY-MM 형식
        return datetime.strptime(date_str, '%y-%m')
    except ValueError:
        pass

    try:
        # YYYY.M 형식
        return datetime.strptime(date_str, '%Y.%m')
    except ValueError:
        pass

    try:
        # YY.M 형식
        return datetime.strptime(date_str, '%y.%m')
    except ValueError:
        pass

    # 모든 형식에 맞지 않을 경우 None 반환
    return None

def contains_unit(string):
    pattern = r'[만천백십]'
    return bool(re.search(pattern, string))
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))


current_page=1


url = 'https://www.musinsa.com/ranking/best?period=month_3&mainCategory=001&page='+current_page
driver.get(url)

# 페이지가 완전히 로드될 때까지 대기 (예: main 태그가 로드될 때까지)
try:
    # 예시로 main 태그가 로드될 때까지 최대 10초 대기
    element = WebDriverWait(driver, 2).until(
        EC.presence_of_element_located((By.TAG_NAME, 'main'))
    )

    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
    # 스크롤 다운
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # 페이지 로드를 위해 충분한 시간을 기다림
        time.sleep(1)

        # 현재 페이지 높이
        new_height = driver.execute_script("return document.body.scrollHeight")

    # 추가적인 로드가 없을 때까지 반복
        if new_height == last_height:
            break
        last_height = new_height
    

    
    
    page_source = driver.page_source

    # BeautifulSoup으로 HTML 파싱
    soup = BeautifulSoup(page_source, 'html.parser')

    # main 태그 내용 추출
    main_content = soup.find('main')
    li_list=main_content.find('ul').find_all('li')
    goods_list=[]
    for li in li_list:
        goods_list.append(li['data-goodsno'])
    
    
    
    
    
    
    
    
    for good in goods_list:
        target_url = 'https://www.musinsa.com/app/goods/'+good
        driver.get(target_url)
        # 예시로 main 태그가 로드될 때까지 최대 10초 대기
        element = WebDriverWait(driver, 3).until(
            EC.presence_of_element_located((By.TAG_NAME, 'main'))
        )
        
        page_source = driver.page_source
        # BeautifulSoup으로 HTML 파싱
        soup = BeautifulSoup(page_source, 'html.parser')
        extract_data(soup, good)
finally:
    driver.quit()
```
